# Testing the Artifact 

Here, we are doing some basic testing of the artifact to ensure there aren't any red flags. I plan to: 

1. Ensure there isn't any missing data 
2. Ensure that there are numbers (not other data types/strings) 
3. Print the max and min value to ensure they look about as expected 

```python
from vivarium import Artifact 
import pandas as pd, numpy as np, os 
```

```python
art = Artifact('/ihme/homes/vionma5/vivarium_gates_nutrition_optimization/test_artifacts/nigeria.hdf')
```

```python
def too_many_draws(key): 
    artifact = art.load(key)
    if 'draw_672' in artifact.columns:
        print(f'{key} includes too many draws!')
    return

def too_many_years(key):
    artifact = art.load(key).reset_index()
    if 'year_start' in artifact.columns: 
        if 2000 in artifact.year_start: 
            print(f'{key} includes too many years')

def check_missingness(key):
    artifact = art.load(key)
    if artifact.isnull().values.any(): 
        print(f'{key} includes missing data')
    return


def check_numbers(key):
    artifact = art.load(key)
    if pd.DataFrame(artifact.dtypes != 'float64').values.any():
        print(f'{key} contains non-float type data')
    return 

def print_max_min(key):
    artifact = art.load(key).reset_index()
    maxi = artifact[[col for col in artifact if col.startswith("draw_")]].max()
    maxi = maxi.max()
    mini = artifact[[col for col in artifact if col.startswith("draw_")]].min()
    mini = mini.min()
    print(f'{key} has a max of {maxi} and a min of {mini}')
    return
```
